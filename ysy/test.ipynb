⭐목표 정의

계약일의 흐름에 따라 부동산 실거래가가 어떻게 변하는지 자치구명, 법정동명 차원에서 분석

```python
#계약일을 datetime형식으로 변환

df['계약일'] = pd.to_datetime(df['계약일'], format= '%Y%m%d')
print(df['계약일'].head())
```

```python
#특정 자치구만 추출하여 변화 확인하기

df_target = df[df['자치구명'].isin(['강남구', '서초구', '송파구'])].copy()
trend_individual = df_target.groupby(['계약일','자치구명'])['물건금액'].mean().unstack()
trend_total = df_target.groupby(['계약일'])['물건금액'].mean()

print(trend_individual)
print(trend_total)
```

```python
import plotly.express as px
# 1. 데이터 준비 (아까처럼 날짜 변환이 되어있어야 합니다.)
df['계약일'] = pd.to_datetime(df['계약일'], format='%Y%m%d')
# 2. 강남 3구 데이터만 가져오기
target_gu = ['강남구', '서초구', '송파구']
df_filtered = df[df['자치구명'].isin(target_gu)].copy()
# 3. 그래프 그리기 (이 명령어가 핵심입니다!)
fig = px.line(df_filtered, 
              x='계약일', 
              y='물건금액', 
              color='법정동명',      # 선의 색깔은 동별로 구분 (이게 가장 세부 단위!)
              facet_col='자치구명',  # 자치구별로 그래프 칸을 옆으로 나눈다 (Falldown 구조)
              title='강남 3구 자치구 및 법정동별 가격 추이',
              labels={'물건금액': '매매가(만원)', '계약일': '날짜'},
              hover_data=['법정동명']) # 마우스 올리면 동 이름 표시
# 그래프 출력
fig.show()
```
