# 1️⃣ 데이터 로드 및 병합

### 💬 대사

> "저희 분석에 사용한 데이터는 **서울 열린데이터광장**에서 수집한 **2022년** 데이터입니다.
>
> 총 **4개의 데이터**를 사용했는데요,
> - 편의점 매출 데이터
> - 점포수 데이터  
> - 유동인구 데이터
> - 상권유형 데이터
>
> 이 데이터들을 **행정동 코드**를 기준으로 병합했고,
> 최종적으로 **약 1,635개**의 분석 데이터를 확보했습니다.
>
> **결측치는 0%**로, 데이터 품질에 문제가 없음을 확인했습니다."

### 📌 핵심 숫자
- 데이터 수: **1,635개**
- 행정동 수: **410개**
- 분기 수: **4개** (2022년 1~4분기)
- 결측치: **0%**

# 2️⃣ 상관분석

### ❓ 핵심 질문
**"유동인구가 많으면 매출이 높을까?"**

### 💬 대사

> "먼저 **상관분석**을 통해 변수들 간의 관계를 확인했습니다.
>
> 상관분석이란 두 변수가 얼마나 같이 움직이는지를 **-1에서 1 사이의 숫자**로 나타낸 것입니다.
>
> 분석 결과, 매출과 가장 높은 상관관계를 보인 변수는 **점포수(r=0.798)**였습니다.
>
> 유동인구와 매출의 상관계수는 **r=0.502**로 중간 정도의 양의 상관관계를 보였습니다.
>
> 이는 **점포수가 많은 지역**, 즉 상권이 발달한 지역일수록 편의점 매출이 높다는 것을 의미합니다.
>
> 히트맵을 보시면 빨간색일수록 양의 상관관계, 파란색일수록 음의 상관관계를 나타냅니다."

### 📌 핵심 숫자
- 매출-점포수: **r=0.798** (강한 양의 상관) ⭐
- 매출-발달상권: **r=0.636** (중간~강한 상관)
- 매출-유동인구: **r=0.502** (중간 상관)
- 매출-골목상권: **r=0.202** (약한 상관)

# 3️⃣ 다중회귀분석

### ❓ 핵심 질문
**"유동인구, 점포수, 상권유형 중 뭐가 제일 영향 커?"**

### 💬 대사

> "다음으로 **다중회귀분석**을 실시했습니다.
>
> 다중회귀분석이란 **여러 개의 독립변수**가 하나의 종속변수에 미치는 영향을 **동시에** 분석하는 방법입니다.
>
> 저희는 종속변수를 **매출**로, 독립변수를 **유동인구, 점포수, 골목상권, 발달상권, 전통시장**으로 설정했습니다.
>
> 분석 결과, 모델의 설명력(R²)은 약 **65~70%**였습니다.
> 이는 저희 모델이 매출 변동의 65~70%를 설명할 수 있다는 뜻입니다.
>
> 유의한 변수는 **점포수와 발달상권**이었고,
> 특히 점포수가 1개 증가할 때 매출이 약 **3~4억원** 증가하는 것으로 나타났습니다.
>
> 참고로 **골목상권의 계수가 음수(-)**인 이유는,
> 골목상권이 주거지역 근처의 작은 상권이라 **상대적으로 매출이 낮기 때문**입니다."

### 📌 핵심 숫자
- R² (설명력): **약 65~70%**
- 유의한 변수: **점포수, 발달상권**
- 점포수 1개 증가 → 매출 **+3~4억원**

# 4️⃣ VIF 다중공선성 검정

### ❓ 핵심 질문
**"독립변수들끼리 겹치는 거 없어?"**

### 💬 대사

> "회귀분석의 **신뢰성**을 확인하기 위해 **다중공선성 검정**을 실시했습니다.
>
> 다중공선성이란 독립변수들끼리 서로 **너무 높은 상관관계**를 가지는 것을 말합니다.
> 이게 심하면 회귀계수를 **믿을 수 없게** 됩니다.
>
> **VIF**라는 지표로 측정하는데,
> - VIF > 10: 심각한 문제
> - VIF > 5: 주의 필요
> - VIF < 5: 문제 없음
>
> 저희 분석 결과, 모든 변수의 VIF가 **5 미만**으로
> **다중공선성 문제가 없음**을 확인했습니다."

### 📌 핵심 숫자
- 모든 VIF: **5 미만** ✅
- 판정: **다중공선성 문제 없음**

# 5️⃣ 잔차분석

### ❓ 핵심 질문
**"우리 모델 믿을 만해?"**

### 💬 대사

> "다음으로 **잔차분석**을 통해 회귀모델의 가정이 충족되는지 확인했습니다.
>
> **잔차**란 실제값과 예측값의 차이입니다.
> 좋은 모델이라면 잔차가 **정규분포**를 따르고, **분산이 일정**해야 합니다.
>
> **정규성 검정**(Shapiro-Wilk)과 **등분산성 검정**(Breusch-Pagan)을 실시했습니다.
>
> 그래프를 보시면, **Q-Q Plot**에서 점들이 대각선에 가깝게 분포하여
> 정규성이 어느 정도 충족됨을 확인할 수 있습니다."

### 📌 핵심 포인트
- 정규성: Q-Q Plot 대각선 확인
- 등분산성: 잔차 vs 적합값 그래프 확인
- Shapiro-Wilk 검정, Breusch-Pagan 검정 실시

# 6️⃣ 더빈-왓슨 검정

### ❓ 핵심 질문
**"분기별 데이터가 서로 영향 안 줘?"**

### 💬 대사

> "**더빈-왓슨 검정**은 잔차 간에 **자기상관**이 있는지 확인하는 검정입니다.
>
> 자기상관이란 이전 시점의 잔차가 다음 시점의 잔차에 영향을 주는 것을 말합니다.
> **시계열 데이터**에서 중요한 검정입니다.
>
> 더빈-왓슨 통계량은 **0에서 4 사이**의 값을 가지며,
> **2에 가까우면** 자기상관이 없다고 판단합니다.
> - DW < 1.5: 양의 자기상관
> - DW ≈ 2: 자기상관 없음 ✅
> - DW > 2.5: 음의 자기상관
>
> 저희 분석 결과, DW 통계량은 약 **1.8**로
> 2에 가까워 **자기상관이 없음**을 확인했습니다."

### 📌 핵심 숫자
- DW 통계량: **약 1.8** (2에 가까움)
- 판정: **자기상관 없음** ✅

# 7️⃣ Cook's Distance

### ❓ 핵심 질문
**"결과 왜곡하는 이상한 데이터 있어?"**

### 💬 대사

> "**Cook's Distance**는 회귀분석 결과에 **과도한 영향**을 미치는 이상치를 탐지하는 방법입니다.
>
> 쉽게 말해, 특정 데이터 하나를 빼면 **결과가 크게 달라지는지** 확인하는 것입니다.
>
> 분석 결과, 영향력 있는 이상치는 전체의 약 **5% 내외**였습니다.
>
> 이 이상치들은 주로 **강남, 명동** 등 매출이 극단적으로 높은 지역에서 나타났습니다.
>
> 다만 Cook's D가 **1을 넘는** 심각한 이상치는 없어서
> 별도 제거 없이 분석을 진행했습니다."

### 📌 핵심 숫자
- 영향력 이상치: **약 5%**
- Cook's D > 1인 관측치: **없음**
- 주로 강남, 명동 등 고매출 지역

# 8️⃣ ANOVA 분산분석

### ❓ 핵심 질문
**"골목상권 vs 발달상권 vs 전통시장, 매출 차이 있어?"**

### 💬 대사

> "**ANOVA 분산분석**을 통해 상권유형별로 매출에 차이가 있는지 검정했습니다.
>
> ANOVA는 **3개 이상의 그룹** 간 평균 차이를 검정하는 방법입니다.
>
> 저희는 상권유형을 **골목상권, 발달상권, 전통시장, 기타**로 구분했습니다.
>
> 분석 결과,
> - F-통계량: **48.00**
> - p-value: **0.001보다 훨씬 작음**
>
> 상권유형별로 매출에 **통계적으로 유의한 차이**가 있음을 확인했습니다.
>
> 박스플롯을 보시면
> - **발달상권**: 평균 **62.4억원** (가장 높음)
> - **전통시장**: 평균 **28.7억원**
> - **골목상권**: 평균 **26.0억원**
>
> 발달상권이 압도적으로 높은 것을 확인할 수 있습니다."

### 📌 핵심 숫자
- F-통계량: **48.00** ⭐
- p-value: **< 0.001** (매우 유의)
- 발달상권: **62.4억원**
- 전통시장: **28.7억원**
- 골목상권: **26.0억원**

# 9️⃣ Tukey 사후검정

### ❓ 핵심 질문
**"정확히 어디랑 어디가 차이나?"**

### 💬 대사

> "ANOVA에서 차이가 있다는 것은 확인했지만,
> **정확히 어떤 그룹 간에** 차이가 있는지는 알 수 없습니다.
>
> 이를 확인하기 위해 **Tukey 사후검정**을 실시했습니다.
>
> Tukey 검정은 **모든 그룹 쌍**을 비교하여 어디서 차이가 나는지 알려줍니다.
>
> 분석 결과,
> - **발달상권 vs 골목상권**: 차이 **36.4억원** → **유의함** ✅
> - **발달상권 vs 전통시장**: 차이 **33.7억원** → **유의함** ✅
> - **골목상권 vs 전통시장**: 차이 **2.7억원** → 유의하지 않음
>
> 결론적으로 매출 순위는
> # **발달상권(62억) >> 전통시장(29억) ≈ 골목상권(26억)**
> 입니다."

### 📌 핵심 숫자
- 발달 vs 골목: **36.4억원 차이** (유의) ⭐
- 발달 vs 전통: **33.7억원 차이** (유의)
- 골목 vs 전통: 차이 없음

# 🔟 변수선택법 (Stepwise)

### ❓ 핵심 질문
**"최적의 변수 조합은 뭐야?"**

### 💬 대사

> "**변수선택법**을 통해 가장 설명력이 높은 변수 조합을 찾았습니다.
>
> 5개의 독립변수로 만들 수 있는 **모든 조합(31가지)**의 Adjusted R²를 비교했습니다.
>
> **Adjusted R²**는 변수 개수를 고려한 설명력으로,
> 불필요한 변수가 추가되면 오히려 낮아집니다.
>
> 분석 결과, 최적의 변수 조합은
> **유동인구, 점포수, 골목상권, 발달상권, 전통시장** 전체였습니다.
>
> 그래프를 보시면 변수가 많아질수록 설명력이 높아지는 것을 확인할 수 있습니다."

### 📌 핵심 포인트
- 최적 조합: **5개 변수 전체**
- 변수 추가할수록 설명력 증가

# 1️⃣1️⃣ 회귀계수 신뢰구간

### ❓ 핵심 질문
**"회귀계수가 얼마나 확실해?"**

### 💬 대사

> "마지막으로 회귀계수의 **95% 신뢰구간**을 확인했습니다.
>
> 신뢰구간이란 '진짜 회귀계수가 이 범위 안에 있을 확률이 95%'라는 뜻입니다.
>
> 중요한 점은, **신뢰구간에 0이 포함되면** 그 변수는 **유의하지 않다**는 것입니다.
>
> 그래프를 보시면
> - **점포수**와 **발달상권**의 신뢰구간은 **0과 겹치지 않아** 유의함
> - 일부 변수는 0을 포함하여 유의하지 않음
>
> 을 확인할 수 있습니다."

### 📌 핵심 포인트
- 신뢰구간에 0 포함 안 되면 → **유의함**
- 점포수, 발달상권: **유의함** ✅

# 📋 Q&A 예상 질문

### Q1. 왜 골목상권 회귀계수가 음수인가요?
> "골목상권은 주거지역 근처의 작은 상권이라 유동인구가 적고, 상대적으로 매출이 낮습니다.
> 다른 변수(점포수, 발달상권)를 통제했을 때, 골목상권이 많을수록 오히려 매출이 낮아지는 패턴이 나타납니다."

### Q2. R²가 70%면 나머지 30%는 뭔가요?
> "나머지 30%는 저희 모델에 포함되지 않은 다른 요인들입니다.
> 예를 들어 편의점 브랜드, 운영 시간, 경쟁 업체 수, 주차 편의성 등이 영향을 줄 수 있습니다."

### Q3. 발달상권이 매출이 높은 이유는?
> "발달상권은 주로 역세권, 번화가 등 유동인구가 많고 상업 활동이 활발한 지역입니다.
> 사람이 많이 모이니 자연스럽게 매출도 높아집니다."
